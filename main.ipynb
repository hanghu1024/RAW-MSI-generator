### generate TICnorm images

```python
#============================
# process data
#============================
pixels_TIC = pixels[:,0].reshape((-1,1))
pixels_feature = pixels[:,1:]
pixels_feature_TICnorm = pixels_feature/pixels_TIC

# convert to imgs
imgs_raw = pixels_to_imgs(pixels_feature, NumLine, NumSpePerLine)
imgs_TICnorm = pixels_to_imgs(pixels_feature_TICnorm, NumLine, NumSpePerLine)
```

```python
#============================
# plot.show
# TIC normalization
#============================
# spike handling helps with the visualization
scale = 0.999

for i in tqdm(range(imgs_TICnorm.shape[0])):     # edit
    img = imgs_TICnorm[i]
    thre = np.quantile(img, scale)
    title = 'm/z:' + str(round(MassList[i][0], 4)) 
    plt.figure(figsize=(6,6))
    plt.imshow(img, cmap='hot', aspect=20, vmax=thre) #, interpolation='none')
    plt.title(title)
    plt.xticks([])
    plt.yticks([])
    plt.colorbar()
    
    plt.show()
    plt.close()
    plt.clf()
    
    #break
```

## 3. example for QE SIM (another set of defs)

### configurations (MassList is required to include filter info)

```python
# number of line scan folders
NumLine = 37                                                                # edit number of lines

# specify .raw file dir
NameBody = "D:\\MSI-generator-v1.2 example datasets\\QE_SIM\\Undosed_Kidney_Slide22_Sec3_"       # edit
NamePost = ".raw"                                                           # edit

# specify mass list file dir
MassList_dir = "D:\\MSI-generator-v1.2 example datasets\\QE_SIM\\MassList.xlsx"                    # edit

# mass window
MassTolerance = 10

# Locate the extraction output folder. 
MSI_data_output = "D:\\MSI-generator-v1.2 example datasets\\QE_BB\\output"                    # edit

# import mass list
df_MassList = pd.read_excel(MassList_dir)
MassList_filters = df_MassList['filters'].values
MassList_mz = df_MassList['m/z'].values
```

```python
# process filter info
filters_info = get_filters_info(NameBody, NamePost, 1)
PeakCountsPerFilter, mzsPerFilter = get_PeakCountsPerFilter(filters_info, MassList_filters, MassList_mz)
ScansPerFilter = get_ScansPerFilter(NameBody, NamePost, NumLine, filters_info)

print('\nfilters in the file:\n{}'.format(filters_info))
print('\nCounts in each filter:\n{}'.format(PeakCountsPerFilter))
print('\nmzs in each filter:\n{}'.format(mzsPerFilter))
print('\nScans in each filter at each line:\n{}'.format(ScansPerFilter))
```

```python
# define total acquisition time to align pixels
AcqTime = 4.5
NumSpePerLine = 250
TimeStamps_aligned = np.linspace(0, AcqTime, NumSpePerLine)
```

```python
# pixel intensity extraction
pixels_total = get_pixels_values_filter(NumLine, TimeStamps_aligned, filters_info, PeakCountsPerFilter, mzsPerFilter, ScansPerFilter)
```

```python
#===========================================
# save/load file
#===========================================
## save file
#Save_dir = MSI_data_output+'\\pixels.npy'
#np.save(Save_dir, pixels)

## load file
#pixels = np.load('D:...\\pixels.npy')
```

### generate TICnorm images

```python
# data process for each filter, pixel level
pixels_TIC_total = []
pixels_feature_total = []
pixels_feature_TICnorm_total = []

for i in range(filters_info[0].shape[0]):
    pixels = pixels_total[i]
    pixels_TIC = pixels[:, 0].reshape((-1, 1))                  # each filter has a TIC
    pixels_feature = pixels[:, 1:]
    pixels_feature_TICnorm = pixels_feature/pixels_TIC
    
    # accumulate
    pixels_TIC_total.append(pixels_TIC)
    pixels_feature_total.append(pixels_feature)
    pixels_feature_TICnorm_total.append(pixels_feature_TICnorm)
    
# convert to imgs
imgs_raw_total = []
imgs_TICnorm_total = []

for i in range(filters_info[0].shape[0]):
    pixels_feature = pixels_feature_total[i]
    pixels_feature_TICnorm = pixels_feature_TICnorm_total[i]
    
    imgs_raw = pixels_to_imgs(pixels_feature, NumLine, NumSpePerLine)
    imgs_TICnorm = pixels_to_imgs(pixels_feature_TICnorm, NumLine, NumSpePerLine)
   
    # accumulate
    imgs_raw_total.append(imgs_raw)
    imgs_TICnorm_total.append(imgs_TICnorm)
```

```python
# generate TIC imgs
scale = 0.999

for i in range(filters_info[0].shape[0]):
    for j in range(len(mzsPerFilter[i])):
        img = imgs_TICnorm_total[i][j]
        thre = np.quantile(img, scale)
        title = 'TICnorm img  ' 'mz ' + str(round(mzsPerFilter[i][j], 4))
        
        plt.figure(figsize=(6,4))
        plt.imshow(img, cmap='hot', aspect=5, vmax=thre)
        plt.title(title)
        plt.xticks([])
        plt.yticks([])
        plt.colorbar()

        plt.show()
        plt.close()
        plt.clf()

        #break
```
